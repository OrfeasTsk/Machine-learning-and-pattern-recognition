## **Ερώτημα 2: Υπολογισμός παραγώγων**

Χρήσιμες ιδιότητες **διανυσμάτων κολώνας** (column vectors): <br> <br>
$$
\begin{align*}
&\frac {\partial \mathbf {x} }{\partial \mathbf {x} } = \mathbf {I} \tag{i} \\~\\
&\frac {\partial \mathbf {b} ^{\top }\mathbf {x} }{\partial \mathbf {x} }=\frac {\partial \mathbf {x} ^{\top }\mathbf {b} }{\partial \mathbf {x} }= \mathbf{b} \tag{ii} \\~\\
&\frac {\partial \mathbf {A} \mathbf {x} }{\partial \mathbf {x} } = \mathbf {A} ^{\top } \tag{iii}\\~\\
&\frac {\partial \left(\mathbf {u(x)} \cdot \mathbf {v(x)}\right)}{\partial \mathbf {x} } = \frac {\partial \mathbf {u} ^{\top }\mathbf {v} }{\partial \mathbf {x} } = \frac {\partial \mathbf {u} }{\partial \mathbf {x} }\mathbf {v} + \frac {\partial \mathbf {v} }{\partial \mathbf {x} }\mathbf {u} \quad \text{(product rule)} \tag{iv} 
\end{align*}
$$ <br>

Χρήσιμες ιδιότητες **πινάκων** (matrices): <br> <br>
$$
\begin{align*}
&\mathbf{U}^{\top^\top} = \mathbf{U} \tag{v}  \\~\\
&\left(\mathbf{U} + \mathbf{V}\right)^\top = \mathbf{U}^\top + \mathbf{V}^\top \tag{vi} \\~\\
&\left(\mathbf{UV}\right)^\top = \mathbf{V}^\top\mathbf{U}^\top \tag{vii}  \\~\\
&\text{tr}\left(\mathbf{U} + \mathbf{V}\right) = \text{tr}\left(\mathbf{U}\right) + \text{tr}\left(\mathbf{V}\right) \tag{viii} \\~\\
&\|\mathbf{U}\|^2_F = \text{tr}\left(\mathbf{U}^{*}\mathbf{U}\right) = \text{tr}\left(\mathbf{U}^{\top}\mathbf{U}\right) \text{(for real matrices)} \tag{ix} \\~\\
&\frac {\partial \text{tr}\left(\mathbf{U}\mathbf{X}\mathbf{V}\right)}{\partial \mathbf {X} } = \mathbf{U}^\top\mathbf{V}^\top \tag{x} \\~\\
&\frac {\partial \text{tr}\left(\mathbf{U}\mathbf{X}^\top\mathbf{V}\right)}{\partial \mathbf {X} } = \mathbf{V}\mathbf{U} \tag{xi} \\~\\
&\frac {\partial \text{tr}\left(\mathbf{V}^\top \mathbf{X}^\top \mathbf{UXV}\right)}{\partial \mathbf {X} } = \mathbf{U}^\top \mathbf{XVV}^\top + \mathbf{UXVV}^\top \tag{xii} \\~\\
\end{align*}
$$

Οι παραπάνω ιδιότητες έχουν παρθεί από το [The Matrix Cookbook](https://www.math.uwaterloo.ca/~hwolkowi/matrixcookbook.pdf) και τη [Wikipedia](https://en.wikipedia.org/wiki/Matrix_calculus)

#### **2.1** *Έστω $f(\mathbf{x}) = \mathbf{x}^T\mathbf{Ax} + \mathbf{b}^T\mathbf{x}$, μια συνάρτηση d μεταβλητών, με $\mathbf{A}$ συμμετρικό πίνακα (δηλ. $\mathbf{A} = \mathbf{A}^T$) διάστασης d x d, και $\mathbf{b} \in \mathbb{R}^d $. Να υπολογίσετε αναλυτικά βήμα προς βήμα με το χέρι, την παράγωγο της $f(\mathbf{x})$ ως προς το διάνυσμα $\mathbf{x} \in \mathbb{R}^d$ παραθέτωντας όλα τα ενδιάμεσα βήματα.*

$$
\mathbf{A} = \begin{bmatrix}
    a_{11} & a_{12} & … & a_{1d-1} & a_{1d} \\
    a_{21} & a_{22} & … & a_{2d-1} & a_{2d} \\
    &  & \vdots & \\
    a_{d-11} & a_{d-12} & … & a_{d-1d-1} & a_{d-1d}\\
    a_{d1} & a_{d2} & … & a_{dd-1} & a_{dd}
\end{bmatrix}, \quad  
\mathbf{x} = \begin{bmatrix}
    x_1 \\
    x_2 \\
    \vdots\\
    x_{d-1} \\
    x_{d}
\end{bmatrix}, \quad 
\mathbf{b} = \begin{bmatrix}
    b_1 \\
    b_2 \\
    \vdots\\
    b_{d-1} \\
    b_{d}
\end{bmatrix}
$$ <br> <br>

Έστω $g(\mathbf{x}) = \mathbf{x}^T\mathbf{Ax}$ και $h(\mathbf{x}) = \mathbf{b}^T\mathbf{x}$, η $f(\mathbf{x})$ μπορεί να γραφτεί ως: <br><br> 

$$f(\mathbf{x}) = g(\mathbf{x}) + h(\mathbf{x}) \tag{1}$$ <br>

Λόγω της γραμμικής ιδιότητας της παραγώγου η $\nabla_\mathbf{x}f(\mathbf{x})$ μπορεί να γραφτεί ως το άθροισμα των επιμέρους παραγώγων των δύο όρων: <br><br>

$$\nabla_\mathbf{x}f(\mathbf{x}) = \nabla_\mathbf{x}g(\mathbf{x}) + \nabla_\mathbf{x}h(\mathbf{x}) \tag{2}$$ <br>

α) Υπολογίσμός $\nabla_\mathbf{x}g(\mathbf{x})$ (το αποτέλεσμα υπαρχεί και στο Matrix Cookbook)
*   Επίλυση με την χρήση των ιδιοτήτων <br><br>
Η $g(\mathbf{x})$ μπορεί να γραφτεί ως το εσωτερικό γινόμενο δύο επιμέρους διανυσματικών συναρτήσεων $\mathbf{u}(\mathbf{x}) = \mathbf{x}$ και $\mathbf{v}(\mathbf{x}) = \mathbf{Ax}$ οπότε: <br><br>
$$g(\mathbf{x}) = \mathbf{u}^T \mathbf{v} \tag{3}$$ <br>
Από την ιδιότητα $(iv)$ προκύπτει: <br> <br>
$$
\begin{align*}
&\nabla_\mathbf{x}g(\mathbf{x}) = {\frac {\partial \mathbf {u} ^T\mathbf {v} }{\partial \mathbf {x} }} \quad ⇔ \\~\\
&\nabla_\mathbf{x}g(\mathbf{x}) = {\frac {\partial \mathbf {u} }{\partial \mathbf {x} }}\mathbf {v} + {\frac {\partial \mathbf {v} }{\partial \mathbf {x} }}\mathbf {u} \quad ⇔  \\~\\
&\nabla_\mathbf{x}g(\mathbf{x}) = {\frac {\partial \mathbf {x} }{\partial \mathbf {x} }}\mathbf {Ax} + {\frac {\partial \mathbf {Ax} }{\partial \mathbf {x} }}\mathbf {x} \tag{4}
\end{align*}
$$<br>
Από τις ιδιότητες $(i)$ και $(iii)$ προκύπτει: <br><br>
$$
\begin{align*}
&\nabla_\mathbf{x}g(\mathbf{x}) = \mathbf{I} \mathbf{Ax} + \mathbf{A}^T\mathbf{x} \quad ⇔  \\~\\
&\nabla_\mathbf{x}g(\mathbf{x}) = \mathbf{Ax} + \mathbf{A}^T\mathbf{x} \quad ⇔  \\~\\
&\nabla_\mathbf{x}g(\mathbf{x}) = (\mathbf{A} + \mathbf{A}^T)\mathbf{x} \tag{5}
\end{align*}
$$<br>
Αφού ο $\mathbf{A}$ είναι συμμετρικός $\mathbf{A} = \mathbf{A}^T$: <br> <br>
$$ \nabla_\mathbf{x}g(\mathbf{x}) = 2\mathbf{Ax} \tag{6} $$
<br><br>

*  Επαλήθευση με αναλυτική επίλυση <br> <br>
Η $g(\mathbf{x})$ αποτελεί τετραγωνική μορφή (quadratic form) και γράφεται ως εξής: <br>
$$
\begin{align*}
&g(\mathbf{x}) = \sum_{i=1}^{d} x_i \sum_{j=1}^{d} a_{ij} x_j ⇔ \quad \\~\\
&g(\mathbf{x}) = \sum_{i=1}^{d}\sum_{j=1}^{d} x_i (a_{ij} x_j) \tag{7}
\end{align*}
$$ <br>
Για την εύρεση της $\nabla_\mathbf{x}g(\mathbf{x})$ αρκεί μόνο να υπολογιστεί η παράγωγος της ποσότητας $\sum_{i=1}^{d}\sum_{j=1}^{d}x_i (a_{ij} x_j)$ ως προς κάθε $x_k$ <br> 
Διακρίνονται οι εξής περιπτώσεις:
  *   $i \neq j$ <br>
  $$s = \sum_{i=1}^{d}\sum_{j=1 \\ j\neq i}^{d}x_i (a_{ij} x_j)$$<br>
  Πρακτικά οι μη σταθεροί όροι που λαμβάνονται υπόψη από την παράγωγο προκύπτουν από το φιξάρισμα των $i,j$ στην τιμή $k$ <br>
  Για i = k προκύπτουν όροι της μορφής $x_k(a_{kj}x_j)$ με παράγωγο $\frac{d}{dx_k}x_k(a_{kj}x_j) = a_{kj} x_j \quad ∀j\ne k $ <br>
  Για j = k προκύπτουν όροι της μορφής $(x_ia_{ik})x_k$ με παράγωγο $\frac{d}{dx_k}(x_ia_{ik})x_k = a_{ik}x_i \quad ∀i\ne k $ <br><br> 
  Άρα συνδυάζοντας τα παραπάνω προκύπτει:<br> <br>
  $$\frac{d}{dx_k}(\sum_{i=1}^{d}\sum_{j=1 \\ j\neq i}^{d}x_i (a_{ij} x_j)) = \sum_{j=1 \\ j\neq k}^{d} a_{kj} x_j + \sum_{i=1 \\ i\neq k}^{d}a_{ik}x_i \tag{8}$$ <br>
  
  *   $i = j$ <br><br>
  Πρακτικά ο μη σταθερος όρος που λαμβάνεται υπόψη από την παράγωγο προκύπτει από το φιξάρισμα του $i$ στην τιμή $k$ <br><br>
  $$\frac{d}{dx_k}(\sum_{i=1}^{d}x_i (a_{ii} x_i)) = 2a_{kk} x_k \tag{9}$$
 
  Άρα $$\frac{\partial g(\mathbf {x}) }{\partial x_k} = 2a_{kk} x_k + \sum_{i=1 \\ i\neq k}^{d} a_{ki} x_i + \sum_{i=1 \\ i\neq k}^{d}a_{ik}x_i = 2a_{kk} x_k + \sum_{i=1 \\ i\neq k}^{d} (a_{ki} + a_{ik}) x_i \tag{10}$$ <br>
  Επίσης αφού ο $\mathbf{A}$ είναι συμμετρικός ισχύει πως $a_{ij} = a_{ji}$ και έτσι: <br><br>
  $$\frac{\partial g(\mathbf {x}) }{\partial x_k} = 2a_{kk} x_k + \sum_{i=1 \\ i\neq k}^{d} 2a_{ki} x_i = \sum_{i=1}^{d} 2a_{ki} x_i = 2\sum_{i=1}^{d} a_{ki} x_i \tag{11}$$ <br>

  Έτσι λοιπόν προκύπτει ξανά η $(6)$ ως εξής: $$ \nabla_\mathbf{x}g(\mathbf{x}) =  
  \begin{bmatrix}
    \frac{\partial g(\mathbf {x}) }{\partial x_1} \\
    \frac{\partial g(\mathbf {x}) }{\partial x_2} \\
    \vdots\\
    \frac{\partial g(\mathbf {x}) }{\partial x_{d-1}} \\
    \frac{\partial g(\mathbf {x}) }{\partial x_d}
  \end{bmatrix} = 2 * \begin{bmatrix}
    \sum_{i=1}^{d} a_{1i} x_i \\
    \sum_{i=1}^{d} a_{2i} x_i \\
    \vdots\\
    \sum_{i=1}^{d} a_{d-1i} x_i \\
    \sum_{i=1}^{d} a_{di} x_i
  \end{bmatrix} = 2\mathbf{Ax} $$

β) Υπολογίσμός $\nabla_\mathbf{x}h(\mathbf{x})$
*   Επίλυση με την χρήση των ιδιοτήτων <br><br>
Από την ιδιότητα $(ii)$ προκύπτει: <br> <br>

$$ \nabla_\mathbf{x}h(\mathbf{x}) = {\frac {\partial \mathbf {b} ^{\top }\mathbf {x} }{\partial \mathbf {x} }} = \mathbf{b} \tag{12}$$

*   Επαλήθευση με αναλυτική επίλυση <br> <br>
Η $h(\mathbf{x})$ αποτελεί εσωτερικό γινόμενο (dot product) και γράφεται ως εξής: <br>
$$h(\mathbf{x}) = \sum_{i=1}^{d}b_ix_i$$ <br>
Για την εύρεση της $\nabla_\mathbf{x}g(\mathbf{x})$ αρκεί μόνο να υπολογιστεί η παράγωγος της ποσότητας $\sum_{i=1}^{d}b_ix_i$ ως προς κάθε $x_k$ <br> 
Πρακτικά ο μη σταθερος όρος που λαμβάνεται υπόψη από την παράγωγο προκύπτει από το φιξάρισμα του $i$ στην τιμή $k$ <br><br>
$$\frac{\partial h(\mathbf {x}) }{\partial x_k} = \frac{d}{dx_k}\sum_{i=1}^{d}b_ix_i = b_k $$
Έτσι λοιπόν προκύπτει ξανά η $(12)$ ως εξής: $$ \nabla_\mathbf{x}h(\mathbf{x}) =  
  \begin{bmatrix}
    \frac{\partial h(\mathbf {x}) }{\partial x_1} \\
    \frac{\partial h(\mathbf {x}) }{\partial x_2} \\
    \vdots\\
    \frac{\partial h(\mathbf {x}) }{\partial x_{d-1}} \\
    \frac{\partial h(\mathbf {x}) }{\partial x_d}
  \end{bmatrix} = \begin{bmatrix}
    b_1 \\
    b_2  \\
    \vdots\\
    b_{d-1}\\
    b_{d}
  \end{bmatrix} = \mathbf{b} $$ <br>

Οπότε από την $(2)$, $(6)$ και $(12)$ προκύπτει πως: <br> <br>  
$$\nabla_\mathbf{x}f(\mathbf{x}) = 2\mathbf{Ax} + \mathbf{b}$$

#### **2.2** *Έστω $\mathbf{Α}$, $\mathbf{Β}$, $\mathbf{Χ}$ πίνακες πραγματικών αριθμών διάστασης $d \times d$ με τάξη ($\operatorname {rank}$) ίση με $d$. Nα βρείτε, αναλυτικά και σε κλειστή μορφή (closed form solution) το ολικό ελάχιστο του παρακάτω προβλήματος βελτιστοποίησης: <br><br> $$\underset{\mathbf{X}}{\mathrm{min}}\|\mathbf{A} − \mathbf{X}\mathbf{B}\|^2_F$$ <br> Να εξηγήσετε με λεπτομέρεια όλη τη διαδικασία σκέψης που θα ακολουθήσετε και να παραθέσετε όλα τα ενδιάμεσα βήματα των πράξεων.*

$$
\mathbf{A} = \begin{bmatrix}
    a_{11} & a_{12} & … & a_{1d-1} & a_{1d} \\
    a_{21} & a_{22} & … & a_{2d-1} & a_{2d} \\
    &  & \vdots & \\
    a_{d-11} & a_{d-12} & … & a_{d-1d-1} & a_{d-1d}\\
    a_{d1} & a_{d2} & … & a_{dd-1} & a_{dd}
\end{bmatrix}, \quad 
\mathbf{B} = \begin{bmatrix}
    b_{11} & b_{12} & … & b_{1d-1} & b_{1d} \\
    b_{21} & b_{22} & … & b_{2d-1} & b_{2d} \\
    &  & \vdots & \\
    b_{d-11} & b_{d-12} & … & b_{d-1d-1} & b_{d-1d}\\
    b_{d1} & b_{d2} & … & b_{dd-1} & b_{dd}
\end{bmatrix}, \quad \\~\\
\mathbf{X} = \begin{bmatrix}
    x_{11} & x_{12} & … & x_{1d-1} & x_{1d} \\
    x_{21} & x_{22} & … & x_{2d-1} & x_{2d} \\
    &  & \vdots & \\
    x_{d-11} & x_{d-12} & … & x_{d-1d-1} & x_{d-1d}\\
    x_{d1} & x_{d2} & … & x_{dd-1} & x_{dd}
\end{bmatrix} 
$$ <br>
Σε πρώτο βήμα θα πρέπει να υπολογιστεί η $\nabla_\mathbf{X}f(\mathbf{X})$ όπου <br><br>

$$f(\mathbf{X}) = \|\mathbf{A} − \mathbf{X}\mathbf{B}\|^2_F \tag{1}$$ <br>
*   Επίλυση με την χρήση των ιδιοτήτων <br><br>
Από την ιδιότητα $(ix)$ ισχύει πως:
$$f(\mathbf{X}) = \text{tr}\left(\left(\mathbf{A} − \mathbf{X}\mathbf{B}\right)^T\left(\mathbf{A} − \mathbf{XB}\right)\right) \tag{2}$$ <br>
Εφαρμόζοντας διαδοχικά τις ιδιότητες $(vi)$ και $(vii)$ στον πρώτο παράγοντα έπειτα την επιμεριστική ιδιότητα και τέλος την ιδιότητα $(viii)$ προκύπτει: <br><br>
$$
\begin{align*}
&f(\mathbf{X}) = \text{tr}\left(\left(\mathbf{A}^T − (\mathbf{ΧB})^T\right)\left(\mathbf{A} − \mathbf{XB}\right)\right) \Leftrightarrow\\
&f(\mathbf{X}) = \text{tr}\left(\left(\mathbf{A}^T − \mathbf{B}^T\mathbf{X}^T\right)\left(\mathbf{A} − \mathbf{XB}\right)\right) \Leftrightarrow\\
&f(\mathbf{X}) = \text{tr}\left(\mathbf{A}^T\mathbf{A} - \mathbf{A}^T\mathbf{XB} - \mathbf{B}^T\mathbf{X}^T\mathbf{A} + \mathbf{B}^T\mathbf{X}^T\mathbf{XB}\right) \Leftrightarrow\\
&f(\mathbf{X}) = \text{tr}\left(\mathbf{A}^T\mathbf{A}\right) - \text{tr}\left(\mathbf{A}^T\mathbf{XB}\right) - \text{tr}\left(\mathbf{B}^T\mathbf{X}^T\mathbf{A}\right) + \text{tr}\left(\mathbf{B}^T\mathbf{X}^T\mathbf{XB}\right) \tag{3}\\
\end{align*}
$$ <br>
Έστω $g(\mathbf{X}) = \text{tr}\left(\mathbf{A}^T\mathbf{A}\right)$, $h(\mathbf{X}) = \text{tr}\left(\mathbf{A}^T\mathbf{XB}\right)$, $p(\mathbf{X}) = \text{tr}\left(\mathbf{B}^T\mathbf{X}^T\mathbf{A}\right)$ και $s(\mathbf{X}) = \text{tr}\left(\mathbf{B}^T\mathbf{X}^T\mathbf{XB}\right)$, η $f(\mathbf{X})$ μπορεί να γραφτεί ως: <br><br> 
$$f(\mathbf{X}) = g(\mathbf{X}) - h(\mathbf{X}) - p(\mathbf{X}) + s(\mathbf{X}) \tag{4}$$ <br>
Λόγω της γραμμικής ιδίοτητας της παραγώγου η $\nabla_\mathbf{X}f(\mathbf{X})$ μπορεί να γραφτεί ως το άθροισμα των επιμέρους παραγώγων των τεσσάρων όρων: <br><br>
$$\nabla_\mathbf{X}f(\mathbf{X}) = \nabla_\mathbf{X}g(\mathbf{X}) - \nabla_\mathbf{X}h(\mathbf{X}) - \nabla_\mathbf{X}p(\mathbf{X}) + \nabla_\mathbf{X}s(\mathbf{X}) \tag{5}$$ <br>
  *   $\nabla_\mathbf{X}g(\mathbf{X}) = 0 $ (δεν εξαρτάται από τον $\mathbf{X}$) <br><br>
  *   $\nabla_\mathbf{X}h(\mathbf{X}) = \mathbf{A}^{T^T}\mathbf{B}^T = \mathbf{A}\mathbf{B}^T$ (ιδιότητες $(x)$ και $(v)$) <br><br>
  *   $\nabla_\mathbf{X}p(\mathbf{X}) = \mathbf{A}\mathbf{B}^T$ (ιδιότητα $(xi)$) <br><br>
  *   $\nabla_\mathbf{X}s(\mathbf{X}) = \mathbf{XBB}^T + \mathbf{XBB}^T = 2 \mathbf{XBB}^T$ (ιδιότητα $(xii)$ όπου $\mathbf{U} = \mathbf{I}$) <br> <br>
Από τα παραπάνω προκύπτει:<br> <br>
$$\nabla_\mathbf{X}f(\mathbf{X}) = 2 \mathbf{XBB}^T -\mathbf{A}\mathbf{B}^T - \mathbf{A}\mathbf{B}^T  = 2 \mathbf{XBB}^T -2\mathbf{A}\mathbf{B}^T  = -2\left(\mathbf{A} - \mathbf{XB}\right)\mathbf{B}^T \tag{6}$$  <br> <br>

*   Επαλήθευση με αναλυτική επίλυση <br> <br>
Η ιδίοτητα $(ix)$ μπορεί να γραφτεί αναλυτικά και ως εξής:<br>
$$\|\mathbf{U}\|^2_F = \sum _{i=1}^{m}\sum _{j=1}^{n}|u_{ij}|^{2}\tag{7}$$<br>
Κάθε στοιχείο $i,j$ του πίνακα $\mathbf{A} − \mathbf{X}\mathbf{B}$ μπορεί να γραφτεί αναλυτικά ως εξής: <br> <br>
$$\left(\mathbf{A} − \mathbf{X}\mathbf{B}\right)_{ij} = a_{ij} - \sum_{l=1}^d x_{il}b_{lj} \tag{8}$$ <br><br>
Άρα από $(7)$ και $(8)$: <br> <br>
$$f(\mathbf{X}) = \sum_{i=1}^d\sum_{j=1}^d(a_{ij}-\sum_{l=1}^d x_{il}b_{lj})^2 \tag{9} $$<br><br>
Για την εύρεση της $\nabla_\mathbf{X}f(\mathbf{X})$ αρκεί μόνο να υπολογιστεί η παράγωγος της ποσότητας $ \sum_{i=1}^d\sum_{j=1}^d(a_{ij}-\sum_{l=1}^d x_{il}b_{lj})^2$ ως προς κάθε $x_{km}$ <br> 
Πρακτικά οι μη σταθεροί όροι που λαμβάνονται υπόψη από την παράγωγο προκύπτουν από το φιξάρισμα του $i$ στην τιμή $k$ και του $l$ στην τιμή $m$ <br><br>
$$\frac{d}{dx_{km}}(\sum_{j=1}^d(a_{kj}-x_{km}b_{mj})^2) = -2\sum_{j=1}^d(a_{kj}-x_{km}b_{mj})b_{mj} = -2\sum_{j=1}^d\left(\mathbf{A}_{kj}-\left(\mathbf{XB}\right)_{kj}\right)b_{mj} = -2\sum_{j=1}^d\left(\mathbf{A}-\mathbf{XB}\right)_{kj}b_{mj} $$ <br>
$$  = -2\left(\left(\mathbf{A}-\mathbf{XB}\right)\mathbf{B}^T\right)_{km} \tag{10}$$ <br>
Οπότε προκύπτει ξανά η $(6)$ ως εξής: <br>
$$\nabla_\mathbf{X}f(\mathbf{X}) = \begin{bmatrix}
    \frac{\partial f(\mathbf {X}) }{\partial x_{11}} & \frac{\partial f(\mathbf {X}) }{\partial x_{12}} & … & \frac{\partial f(\mathbf {X}) }{\partial x_{1d-1}} & \frac{\partial f(\mathbf {X}) }{\partial x_{1d}} \\
    \frac{\partial f(\mathbf {X}) }{\partial x_{21}} & \frac{\partial f(\mathbf {X}) }{\partial x_{22}} & … & \frac{\partial f(\mathbf {X}) }{\partial x_{2d-1}} & \frac{\partial f(\mathbf {X}) }{\partial x_{2d}} \\
    &  & \vdots & \\
    \frac{\partial f(\mathbf {X}) }{\partial x_{d-11}} & \frac{\partial f(\mathbf {X}) }{\partial x_{d-12}} & … & \frac{\partial f(\mathbf {X}) }{\partial x_{d-1d-1}} & \frac{\partial f(\mathbf {X}) }{\partial x_{d-1d}} \\
    \frac{\partial f(\mathbf {X}) }{\partial x_{d1}} & \frac{\partial f(\mathbf {X}) }{\partial x_{d2}} & … & \frac{\partial f(\mathbf {X}) }{\partial x_{dd-1}} & \frac{\partial f(\mathbf {X}) }{\partial x_{dd}}
\end{bmatrix} \\~\\ =  -2 \begin{bmatrix}
    \left(\left(\mathbf{A}-\mathbf{XB}\right)\mathbf{B}^T\right)_{11} &  \left(\left(\mathbf{A}-\mathbf{XB}\right)\mathbf{B}^T\right)_{12} & … & \left(\left(\mathbf{A}-\mathbf{XB}\right)\mathbf{B}^T\right)_{1d-1} & \left(\left(\mathbf{A}-\mathbf{XB}\right)\mathbf{B}^T\right)_{1d} \\
    \left(\left(\mathbf{A}-\mathbf{XB}\right)\mathbf{B}^T\right)_{21} &  \left(\left(\mathbf{A}-\mathbf{XB}\right)\mathbf{B}^T\right)_{22} & … & \left(\left(\mathbf{A}-\mathbf{XB}\right)\mathbf{B}^T\right)_{2d-1} & \left(\left(\mathbf{A}-\mathbf{XB}\right)\mathbf{B}^T\right)_{2d} \\
    &  & \vdots & \\
    \left(\left(\mathbf{A}-\mathbf{XB}\right)\mathbf{B}^T\right)_{d-11} &  \left(\left(\mathbf{A}-\mathbf{XB}\right)\mathbf{B}^T\right)_{d-12} & … & \left(\left(\mathbf{A}-\mathbf{XB}\right)\mathbf{B}^T\right)_{d-1d-1} & \left(\left(\mathbf{A}-\mathbf{XB}\right)\mathbf{B}^T\right)_{d-1d} \\
    \left(\left(\mathbf{A}-\mathbf{XB}\right)\mathbf{B}^T\right)_{d1} &  \left(\left(\mathbf{A}-\mathbf{XB}\right)\mathbf{B}^T\right)_{d2} & … & \left(\left(\mathbf{A}-\mathbf{XB}\right)\mathbf{B}^T\right)_{dd-1} & \left(\left(\mathbf{A}-\mathbf{XB}\right)\mathbf{B}^T\right)_{dd} 
\end{bmatrix} \\~\\ 
= -2\left(\mathbf{A} - \mathbf{XB}\right)\mathbf{B}^T
$$ <br>

Αφού λοιπόν βρέθηκε η $\nabla_\mathbf{X}f(\mathbf{X})$ για την εύρεση του $\underset{\mathbf{X}}{\mathrm{min}}f(\mathbf{X})$ θα πρέπει: <br>

$$\nabla_\mathbf{X}f(\mathbf{X}) = 0 \tag{11}$$

Δηλαδή,

$$
\begin{align*}
&-2\left(\mathbf{A} - \mathbf{XB}\right)\mathbf{B}^T = 0 ⇔\\~\\
&\left(\mathbf{A} - \mathbf{XB}\right)\mathbf{B}^T = 0 ⇔\\~\\
&\mathbf{A}\mathbf{B}^T - \mathbf{XBB}^T = 0 ⇔\\~\\
&\mathbf{XBB}^T = \mathbf{A}\mathbf{B}^T \tag{12}\\
\end{align*}
$$ <br><br>
Αφού ο $\mathbf{B}$ έχει $\operatorname {rank} = d$ από τις ιδιότητες του $\operatorname {rank}$ ισχύει πως $\operatorname {rank} (\mathbf{B}^T\mathbf{B})=\operatorname {rank} (\mathbf{B}\mathbf{B}^T)=\operatorname {rank} (\mathbf{B})=\operatorname {rank} (\mathbf{B}^T) = d$ 

Επίσης αφού ο $\mathbf{B}\mathbf{B}^T$ είναι διάστασης $d \times d$ και $\operatorname {rank} (\mathbf{B}\mathbf{B}^T) = d$ σημαίνει πως όλες οι γραμμές είναι γραμμικά ανεξάρτητες μεταξύ τους όπως και οι κολώνες του (είναι **full rank** πίνακας). Δηλαδή καμία γραμμή δεν μπορεί να γραφτεί ως γραμμικος συνδιασμός των υπολοίπων και το ίδιο ισχύει και για τις κολώνες. Από την θεωρία είναι γνωστό πως **κάθε full rank πίνακας είναι και αντιστρέψιμος**.<br><br>
Οπότε υπάρχει ο $(\mathbf{B}\mathbf{B}^T)^{-1}$ και έτσι πολλαπλασίαζοντας τον από δεξιά προκύπτει: <br><br>

$$
\begin{align*}
&\mathbf{XBB}^T(\mathbf{B}\mathbf{B}^T)^{-1} = \mathbf{A}\mathbf{B}^T(\mathbf{B}\mathbf{B}^T)^{-1} ⇔ \\~\\
&\mathbf{XI} = \mathbf{A}\mathbf{B}^T(\mathbf{B}\mathbf{B}^T)^{-1} ⇔ \\~\\
&\mathbf{X} = \mathbf{A}\mathbf{B}^T(\mathbf{B}\mathbf{B}^T)^{-1} \tag{13}\\
\end{align*}
$$ <br><br>

Από την $(13)$ προκέκυψε ένας **κλειστός τύπος** για την εύρεση των τιμών του $\mathbf{X}$ που ελαχιστοποιούν την $f(\mathbf{X})$

## **Ερώτημα 3: Gradient descent**

#### *Ο αλγόριθμος gradient descent είναι η πιο διαδεδομένη μέθοδος βελτιστοποίησης στη μηχανική μάθηση. Σε αυτό το ερώτημα καλείστε να υλοποιήσετε σε Python τον αλγόριθμο gradient descent για την ελαχιστοποίηση συναρτήσεων δύο μεταβλητών. <br> <br> Θεωρήστε τις παρακάτω συναρτήσεις:<br> <br> $$f_1\left([x_1, x_2]^T\right) = f_1\left(x_1, x_2\right) = \left(x_1 − 2\right)^2 + \left(x_2 − 3\right)^2 \\~\\
f_2\left([x_1, x_2]^T\right) = f_2\left(x_1, x_2\right) = \left(1 − \left(x_2 − 3\right)\right)^2 + 20 \left(\left(x_1 + 3\right) − \left(x_2 − 3\right)^2\right)^2 \\ = \left(4 - x_2\right)^2 + 20 \left(3 + x_1 − \left(x_2 − 3\right)^2 \right)^2
$$ <br> Nα υλοποιήσετε σε Python και να εκτελέστε τον αλγόριθμο gradient descent για κάθε συνάρτηση.*

Σημαντική προϋπόθεση για την εκτέλεση του gradient descent αποτελεί η εύρεση της κλίσης της κάθε συνάρτησης. Και οι δύο συναρτήσεις λαμβάνουν ένα διάνυσμα $\mathbf{x}$ οπότε θα πρέπει να βρεθεί η μερική παράγωγος ως προς αυτό το διάνυσμα εισόδου δηλαδή ως προς τα $x_1$ και $x_2$. Για τον υπολογισμό των μερικών παραγώγων γίνεται κυρίως εφαρμογή του κανόνα αλυσίδας/σύνθεσης. <br> <br>

$$
\begin{align*}
&\frac{\partial f_1(\mathbf {x}) }{\partial x_{1}} = 2(x_1 - 2) \\~\\
&\frac{\partial f_1(\mathbf {x}) }{\partial x_{2}} = 2(x_2 - 3) \\~\\
&\frac{\partial f_2(\mathbf {x}) }{\partial x_{1}} = 40 \left(3 + x_1 − \left(x_2 − 3\right)^2 \right) \\~\\
&\frac{\partial f_2(\mathbf {x}) }{\partial x_{2}} = -2\left(4 - x_2\right) + 40 \left(3 + x_1 − \left(x_2 − 3\right)^2 \right)(-2)\left(x_2 - 3\right) = 2\left(\left(x_2 - 4\right) - 40 \left(x_2-3\right) \left(3 + x_1 − \left(x_2 − 3\right)^2 \right)\right) = 2\left(- 1 + \left(x_2 - 3\right) + 40 \left(x_2 - 3\right) \left( \left(x_2 − 3\right)^2 - \left(x_1 + 3 \right)\right)\right) = 2\left(\left(x_2 - 3\right) \left(1 + 40\left(\left(x_2 − 3\right)^2 - \left(x_1 + 3\right)\right)\right) - 1\right) \\~\\~\\ 
&\nabla_\mathbf{x}f_1(\mathbf{x}) = 2\begin{bmatrix}
x_1 - 2 \\ 
x_2 - 3
\end{bmatrix} \\~\\
&\nabla_\mathbf{x}f_2(\mathbf{x}) = 2\begin{bmatrix}
20 \left(3 + x_1 − \left(x_2 − 3\right)^2 \right)\\ 
\left(x_2 - 3\right) \left(1 + 40\left(\left(x_2 − 3\right)^2 - \left(x_1 + 3\right)\right)\right) - 1
\end{bmatrix}
\end{align*}
$$ <br>

Οι τύποι των μερικών παραγώγων που βρέθηκαν για την ${f_1}$ και ${f_2}$ θα υλοποιηθούν μέσω συναρτήσεων της python με το χέρι. Παρόλα αυτά μερικά frameworks όπως το pytorch μπορούν να εκτελέσουν αυτόματο υπολογισμό των μερικών παραγώγων (**autograd**) οποιαδήποτε μαθηματικής συνάρτησης μέσω της python χρήσιμοποιώντας υπολογιστικούς γράφους (**computational graphs**)


```python
import numpy as np

f1 = lambda x: (x[0] - 2) ** 2 + (x[1] - 3) ** 2 
grad_f1 = lambda x: 2 * np.array([x[0] - 2, x[1] - 3])

f2 = lambda x: (4 - x[1]) ** 2 + 20 * (3 + x[0] - (x[1] - 3) ** 2) ** 2
grad_f2 = lambda x: 2 * np.array([20 * (3 + x[0] - (x[1] - 3) ** 2), (x[1] - 3) * (1 + 40 * ((x[1] - 3) ** 2 - (x[0] + 3))) - 1])
```

Στην συνέχεια υλοποιείται η μέθοδος που θα χρησιμοποιηθεί για το gradient descent:

```python
import matplotlib.pyplot as plt

def gradient_descent(f, grad_f, x_init, learning_rate, max_iters = 1000000, epsilon = None, verbose = False):
    
    x_list = [x_init]
    x = x_init
    
    if verbose:
      print(f'Iterations passed: 0, y = f({",".join(map(str, x))}) = {f(x)} (initial value)')

    i = 1 
    while True:

        # Calculation of the new 'x' using the current 'x' and the gradient at that current 'x'
        x = x - learning_rate * grad_f(x)

        if verbose:
          print(f'Iterations passed: {i}, y = f({",".join(map(str, x))}) = {f(x)}')

        x_list.append(x)

        # Checking if the maximum number of iterations has been reached
        if max_iters is not None:
          if i == max_iters:
            print(f'\n Maximum number of iterations has been reached\n')
            break
      
        # Checking if changes/updates are negligible according to a threshold
        if epsilon is not None:
          # Euclidean (l2) distance between the last 'x' and the second to last 'x'
          if np.linalg.norm(x_list[-1] - x_list[-2]) <= epsilon:
            print(f'\n Optimal solution with precision {epsilon} found\n')
            break

        i += 1
    
    print(f'\nSummary: Found x = {x_list[-1]} with y = f({",".join(map(str, x_list[-1]))}) = {f(x_list[-1])} in {i} iterations\n')
    # Plot with function values vs iters
    plt.figure(figsize = (8, 8))
    plt.ylabel('f(x)')
    plt.xlabel('Iterations')
    plt.title('f(x) vs Iterations Graph')
    plt.plot(range(i + 1), list(map(f, x_list)))
    plt.show()

    return x_list
```

#### **3.1** *Να αρχικοποιήσετε τις μεταβλητές με μηδέν. Πρώτα, τρέξτε με ένα σταθερό ποσοστό εκμάθησης (learning rate) ίσο με 0.5. Να εκτελέστε Τ=10 επαναλήψεις και αναφέρετε την τιμή της συνάρτησης στο τέλος κάθε βήματος (επανάληψης). Παρακάτω παρουσιάζεται η διαδικασία αναζήτησης της βέλτιστης λύσης της $f_1$ εκτελώντας gradient descent 10 επαναλήψεων με learning rate 0.5 και αρχικές τιμές $x_1 = 0, x_2 = 0$. Φαίνεται πως από την πρώτη κιόλας επανάληψη βρίσκεται η ελάχιστη τιμή της για $x_1 = 2$ και $x_2 = 3$ δηλαδή έχει **συγκλίνει σε στάσιμο σημείο**

```python
# Initialization of the vector x with 0s (x1 = 0, x2 = 0)
x = np.zeros(2)

max_iters = 10

learning_rate = 0.5

print('Searching for the optimal x of f1(x): \n')
x_list = gradient_descent(f1, grad_f1, x, learning_rate, max_iters = max_iters, verbose = True)
```

Παρακάτω παρουσιάζεται η διαδικασία αναζήτησης της βέλτιστης λύσης της $f_2$ εκτελώντας gradient descent 10 επαναλήψεων με learning rate 0.5 και αρχικές τιμές $x_1 = 0, x_2 = 0$. Φαίνεται πως σε επανάληψη η τιμή της **μεγαλώνει με τεράστιο ρυθμό** με αποτέλεσμα να απομακρύνεται αρκετά από την ελάχιστη τιμή. Οι τιμές αυτές μεγαλώνουν μέχρι την **μέγιστη δυνατή τιμή** (inf) που μπορεί να αναπαρασταθεί και στην περίπτωση αυτή τα αποτελέσματα που προκύπτουν αρχίζουν να γίνονται **απροσδιόριστα** (nan)

```python
x = np.zeros(2)

max_iters = 10

learning_rate = 0.5

print('Searching for the optimal x of f2(x): \n')
x_list = gradient_descent(f2, grad_f2, x, learning_rate, max_iters = max_iters, verbose = True)
```

**Συμπεράσματα για τις διαφορετικές τιμές του learning rate** <br><br>
Μεταξύ των πειραμάτων που έγιναν για τις δύο συναρτήσεις διαπιστώνονται κοινά χαρακτηριστικά. Σε όλα τα πειράματα χρησιμοποιήθηκε η ίδια σταθερά σύγκλισης (precision) με τιμή $10^{-6}$. Αυτό που παρατηρήθηκε σχεδόν σε όλες τις περιπτώσεις είναι πως **όσο αυξάνεται η τιμή του learning rate οι επαναλήψεις του αλγορίθμου μειώνονται σημαντικά** ενώ παράλληλα **η τιμή της συνάρτησης** φαίνεται να **προσεγγίζει περισσότερο και πιο γρήγορα την πραγματική βέλτιστη τιμή** (μπορεί να φανεί και από την πιο απότομη κλίση των γραφημάτων). Δηλαδή οι μικρότερες τιμές learning rate φαίνεται να χρειάζονται μικρότερη σταθερά σύγκλισης για να έχουν την ίδια απόδοση με τις μεγαλύτερες, πράγμα που σημαίνει πως απαιτούν ακόμα περισσότερες επαναλήψεις. Παρόλα αυτά το μεγάλο μειονέκτημα για  **μεγάλες τιμές** του learning rate είναι πως υπάρχει ο κίνδυνος να **αποκλίνει κατά πολύ η τιμή της συνάρτησης από τη βέλτιστη τιμή** ή και να απέχει από εκείνη για κάμποσες επαναλήψεις, είτε επειδή απομακρύνεται συνεχώς (βλ. υποπεριπτώσεις της συνάρτησης $f_2$ για learning rate $10^{-2}$, $10^{-1}$ και $1$) είτε επειδή παραμένει πάντοτε σταθερή χωρίς να σημειώνεται πρόοδος (βλ. υποπεριπτώσεις της συνάρτησης $f_1$ για learning rate $1$). Μια ακόμα ειδική περίπτωση που μπορεί να παρατηρηθεί για τις μεγάλες τιμές είναι πως μπορεί να γίνονται **"ταλαντώσεις" επ άπειρον γύρω από την βέλτιστη λύση** λόγω των μεγάλων βημάτων εμποδίζοντας να επιτευχθεί κάποια στιγμή σύγκλιση (βλ. πρώτη υποπερίπτωση συνάρτησης $f_2$ για learning rate $10^{-2}$). Διακρίνεται δηλαδή κάποιου είδους **trade-off** μεταξύ των μεγάλων και των μικρών τιμών μιας και για μικρές τιμές η σύγκλιση σε στάσιμο σημείο μπορεί μεν να επιτευχθεί αλλά με πολύ περισσότερες επαναλήψεις σε σχέση με τις μεγάλες τιμές που χρειάζονται λιγότερες επαναλήψεις εκτός και αν υπάρξει οποιουδήποτε είδους απόκλιση ή "ταλάντωση". <br> <br>
Μια καλή τεχνική που χρησιμοποιείται συχνά για την εξισσορόπιση των δύο αυτών παραγόντων είναι η **δυναμική προσαρμογή του learning rate** (**learning rate schelduling**) χρησιμοποιώντας κάποια αντίστοιχη μέθοδο προσαρμογής του learning rate ξεκινώντας από μεγαλύτερες τιμές και μειώνοντας σταδιακά τις τιμές αυτές με την πάροδο των επαναλήψεων. Με αυτόν τον τρόπο αυξάνονται οι πιθανότητες σύγκλισης σε στάσιμο σημείο αφού τα βήματα στο τέλος γίνονται πιο μικρά ενώ ταυτόχρονα η σύγκλιση αυτή πραγματοποιείται και πιο γρήγορα μιας και στην αρχή γίνονται κάποια μεγάλα βήματα. Πρακτικά η μέθοδος του gradient descent είναι μια μέθοδος που βασίζεται σε τοπική αναζήτηση (**local search**) όπου εκεί συναντάται κυρίως η ιδέα της δυναμικής προσαρμογής των βημάτων αναζήτησης (π.χ simulated annealing). <br> <br>
Άρα η επιλογή του καλύτερου learning rate διαφέρει ανάλογα με τη συνάρτηση και όπως διαπιστώθηκε είτε χρείαζεται να γίνει πειραματισμός μεταξύ πολλών διαφορετικών τιμών ή ακόμα μπορεί να χρειαστεί να γίνει εφαρμογή κάποιας μεθόδου δυναμικής προσαρμογής των τιμών του learning rate

**Συμπεράσματα για τις διαφορετικές τιμές αρχικοποίησης των** $x_1$ **και** $x_2$ <br> <br>
 
Σε κάθε πείραμα προκύπτει πως **όσο οι τιμές αρχικοποίησης των** $x_1$ **και** $x_2$ **είναι αρκετά κοντά στην βέλτιστη λύση τόσο λιγότερες επαναλήψεις χρείαζονται** από τον αλγόριθμο gradient descent να συγκλίνει. Αυτό είναι αρκετά λογικό μιας και διαισθητικά γίνεται αντιληπτό και από το γεγονός πως αν γινόταν αρχικοποιήση ακριβώς με την τιμή της βέλτιστης λύσης τότε ο αλγόριθμος θα έχει συγκλίνει αμέσως σε στάσιμο σημείο. Επιπλέον ένα άλλο συμπέρασμα που προκύπτει είναι πως **η θέση αρχικοποιήσης της αναζήτησης επηρεάζει αρκετά την σύγκλιση**. Για παράδειγμα, αν η αρχικοποίηση γίνει κοντά στην βέλτιστη λύση υπάρχει μεγαλύτερη πιθανότητα σύγκλισης σε αντίθεση με το να γινόταν μακριά από τη βέλτιστη λύση όπου εκεί η αναζήτηση μπορεί να "χάσει" τον δρόμο και να αποκλίνει από εκείνη (βλ. τρίτη υποπερίπτωση σε όλα τα πειράματα της συνάρτησης $f_2$).
